```python
# Data Analysis Template: Relationship Between GDP and Life Expectancy

# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Load the Data
# Purpose: Import and preview the dataset
file_path = "all_data.csv"  # Adjust the path if needed
data = pd.read_csv(file_path)

# Preview the data
print("First 5 rows of the dataset:")
print(data.head())
print("\nSummary of the dataset:")
print(data.info())
```

```python
# 2. Data Cleaning
# Purpose: Handle missing values and correct data inconsistencies
print("\nChecking for missing values:")
print(data.isnull().sum())

# Handle missing values (example: drop rows with NaN or fill with median)
data.dropna(inplace=True)

# Ensure column names are consistent for ease of use
data.rename(columns={
    "Life expectancy at birth (years)": "Life_Expectancy",
    "GDP": "GDP"
}, inplace=True)

# Check for duplicates
print("\nChecking for duplicate rows:")
print(data.duplicated().sum())
data.drop_duplicates(inplace=True)
```

```python
# 3. Exploratory Data Analysis (EDA)
# Purpose: Understand the data distribution and key patterns

# Distribution of GDP
plt.figure(figsize=(8, 4))
sns.histplot(data['GDP'], bins=30, kde=True, color='blue')
plt.title('Distribution of GDP')
plt.xlabel('GDP')
plt.ylabel('Frequency')
plt.show()

# Distribution of Life Expectancy
plt.figure(figsize=(8, 4))
sns.histplot(data['Life_Expectancy'], bins=30, kde=True, color='green')
plt.title('Distribution of Life Expectancy')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Frequency')
plt.show()

# Relationship between GDP and Life Expectancy
plt.figure(figsize=(10, 6))
sns.scatterplot(x="GDP", y="Life_Expectancy", hue="Country", data=data, palette='tab10')
plt.title('Relationship Between GDP and Life Expectancy')
plt.xlabel('GDP')
plt.ylabel('Life Expectancy at Birth (years)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 4. Statistical Analysis
# Purpose: Validate the relationship between GDP and Life Expectancy

# Compute correlation between GDP and Life Expectancy
correlation = data['GDP'].corr(data['Life_Expectancy'])
print(f"\nCorrelation between GDP and Life Expectancy: {correlation:.2f}")
```
